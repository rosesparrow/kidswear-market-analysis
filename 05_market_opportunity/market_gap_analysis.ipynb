## 2. Rough Gap Sizing & Demand Quantification

Using the cohort-component forecast model (base case, mid-2025 est from ONS historical data + modest migration assumptions) to size the core 11–19 addressable audience more accurately.

```python
# ── Rough Gap Sizing & Demand Quantification ──
# Using cohort-component forecast (base case mid-2025 est from ONS-derived model)

# From forecast: Total 11–19 in 2025 ≈ 7.54M (base scenario)
# Girls share ≈ 49–50% (historical male surplus in 8–19 data)
uk_total_11_19_2025 = 7_543_139          # exact from base forecast mid-2025
uk_girls_11_19 = int(uk_total_11_19_2025 * 0.495)  # conservative ~49.5% female

# Assumptions (conservative, girls-focused for core premium appeal)
premium_penetration_low  = 0.03   # 3% willing/able to pay premium
premium_penetration_high = 0.10   # 10% optimistic (PPP-like virality + social proof)
avg_order_value = 70              # £65–85 midpoint, annual spend proxy

low_sam  = uk_girls_11_19 * premium_penetration_low  * avg_order_value
high_sam = uk_girls_11_19 * premium_penetration_high * avg_order_value

print(f"UK 11–19 population (mid-2025 base forecast est): {uk_total_11_19_2025:,} total")
print(f"  Girls 11–19 approx: {uk_girls_11_19:,} (~49.5% share)")
print(f"Rough SAM estimate (Serviceable Addressable Market, girls-focused):")
print(f"  Low (3% penetration): £{low_sam / 1_000_000:,.1f}M annual potential")
print(f"  High (10% penetration): £{high_sam / 1_000_000:,.1f}M annual potential")
print("\nNotes:")
print(" - Base forecast assumes modest net migration buffer → stable core cohort.")
print(" - Real TAM larger if including boys (full 11–19 ~7.5M), repeat buys, or multi-item orders.")
print(" - Pre-conversion/frequency; upside in optimistic migration scenario (~7.9M+ by 2035).")
```

## 3. Moats, Risks & Opportunity Thesis

### Potential Moats
- TikTok/IG-native virality + founder authenticity (PPP model)
- Premium materials (e.g. Supima cotton, ethical sourcing) harder for fast fashion to replicate
- Matching sets + teen-specific sizing/aesthetics
- Community/loyalty from viral drops

### Key Risks / Barriers
- Fast-fashion fast-follow (copycat designs at lower price)
- Scaling founder-led authenticity beyond early growth
- Teen trend volatility / social algorithm changes
- Supply chain costs for premium fabrics

### Why This Gap Is Attractive
- Proven demand: PPP expansion (hoodies → pyjamas) + search trends
- Social leverage: TikTok underutilized by UK premiums
- Price sweet spot: £55–85 bridges fast fashion and luxury adult
- Timing: Gen Z/Alpha comfortwear normalization post-pandemic

**Overall Thesis**  
The £55–85 premium teen (11–19) loungewear/pyjama segment is a clear, high-potential white space in the UK — with strong social proof from PPP and unmet demand signals.
